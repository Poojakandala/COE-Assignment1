## Assignment 2: Data Filtering & Transformation


- Display rides where CATEGORY == 'Business'.


```python
uber_data[uber_data['CATEGORY*']=='Business']
```

- Show top 5 rides with the longest distance (MILES).

```python
top5=uber_data.sort_values(by='MILES*',ascending=False).head(5)
top5
```

- Replace all missing PURPOSE values with "Not Specified".

```python
uber_data['PURPOSE*'].fillna('Not Specified', inplace=True)
uber_data
```

- Create a new column TRIP_DURATION using END_DATE - START_DATE.


```python
uber_data['START_DATE*'] = pd.to_datetime(uber_data['START_DATE*'], errors='coerce', infer_datetime_format=True)
uber_data['END_DATE*'] = pd.to_datetime(uber_data['END_DATE*'], errors='coerce', infer_datetime_format=True)
uber_data.dtypes
```

```python
uber_data['TRIP_DURATION'] = uber_data['END_DATE*'] - uber_data['START_DATE*']
uber_data
```

- Sort trips by distance in descending order.


```python
rides = uber_data.sort_values(by='MILES*', ascending=False)
rides
```

## Assignment 3: Grouping and Aggregation

- Group by CATEGORY and find average miles per category.

```python
avg_miles=uber_data.groupby('CATEGORY*')['MILES*'].mean()
avg_miles
```


- Find total trips for each PURPOSE.

```python
total_trips=uber_data.groupby('PURPOSE*').size()
total_trips
```
