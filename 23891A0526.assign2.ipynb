## üìä Part 2 ‚Äî City-wise Analysis 

- Find total sales for each city in each month using groupby().

```python
city_total=sales.groupby('city')[['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].sum()
city_total
```

- Find the average monthly sales per city.

```python
city_avg=sales.groupby('city')[['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].mean()
city_avg
```

- The city with the highest average sales.

```python
high_city=city_avg.mean(axis=1).idxmax()
high_city
```

- The city with the lowest total sales.

```python
low_city=city_total.sum(axis=1).idxmin()
low_city
```

- Add a new column Total_Sales = Jan + Feb + Mar + Apr.

```python
sales['Total_sales'] = sales['Jan']+sales['Feb']+sales['Mar']+sales['Apr']
sales
```

- Find the top 3 performing cities based on Total_Sales.

```python
top3=sales.sort_values(by='Total_sales',ascending=False).head(3)
top3['city']
```

## üßÆ Part 4 ‚Äî Aggregation and Analysis
- Using groupby("city"), compute:

- Mean, Max, and Min for each month.

```python
sales_by_city=sales.groupby('city')[['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].agg(['mean','max','min'])
sales_by_city
```

- Find the city with the highest average March sales.

```python
march_avg=sales.groupby('city')['Mar'].mean()
top_city=march_avg.idxmax()
top_city
```

- Calculate the overall average sales per city (across all months).

```python
overall_avg=sales.groupby('city')[['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].mean().mean(axis=1)
overall_avg
```

- Identify the month with the highest overall sales.

```python
sales['MONTH_MAX'] = sales[['Jan','Feb','Mar','Apr','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].idxmax(axis=1)
sales[['store_id','Jan','Feb','Mar','Apr','MONTH_MAX']]
```

- For each city, find the month with maximum total sales.

```python
total_per_month=sales[['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].sum()
month_highest=total_per_month.idxmax()
month_highest
```

- Add a column Average_Sales (average of all months per store).

```python
sales['AVERAGE_SALES']=sales[['Jan','Feb','Mar','Apr','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].mean(axis=1)
sales.head()
```

- Sort the dataset by this column in descending order.

```python
top=sales.sort_values(by='AVERAGE_SALES',ascending=False)
top
```
